Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test,y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=43)

model= LogisticRegression()
model.fit(X_train,y_train)
result = model.score(X_test, y_test)
print(result)
# for random_state=0, score= 0.9590643274853801
# for random_state=100, score= 0.9415204678362573
# for random_state=59, score= 0.935672514619883
# for random_state=42, score= 0.9707602339181286
# for random_state=43, score= 0.9766081871345029
```

Stratified K-fold Cross Validation

```python
from sklearn.model_selection import StratifiedKFold

skfold = StratifiedKFold(n_splits=5)
result = cross_val_score(model,X,y, cv= skfold)
print(np.mean(result))
# for k= 5 , 0.943766495885732
```

LOOCV - Leave One OUt CV

```python
from sklearn.model_selection import LeaveOneOut

lc = LeaveOneOut()
result = cross_val_score(model,X,y, cv= lc)
print(np.mean(result))
# 0.9472759226713533
```
